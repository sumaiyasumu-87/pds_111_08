```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


final_marks = merged_df[['Final Marks']]


kmeans = KMeans(n_clusters=3, random_state=42)
merged_df['Cluster'] = kmeans.fit_predict(final_marks)

plt.figure(figsize=(8, 6))
plt.scatter(merged_df['Student Name'], merged_df['Final Marks'], c=merged_df['Cluster'], cmap='viridis')
plt.xlabel('Student Name')
plt.ylabel('Final Marks')
plt.title('Clustering of Final Marks')
plt.xticks(rotation=90)
plt.colorbar(label='Cluster')
plt.tight_layout()
plt.show()
```
